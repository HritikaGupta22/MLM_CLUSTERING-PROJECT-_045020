# Data Description
The dataset provides comprehensive information on significant earthquakes worldwide since 1900, with a magnitude of 5 or above. The dataset includes earthquakes from all regions of the world, from the most seismically active regions to less active regions. Geographical clustering can be performed based on seismic activity patterns worldwide, focusing on regions prone to significant earthquakes. This type of clustering aims to identify similarities and differences in earthquake occurrences across different regions, aiding in understanding seismic activity trends and potentially informing disaster preparedness and mitigation strategies.

    1. time: Time of the earthquake in milliseconds.
    2. latitude: Latitude of the earthquake's epicenter in decimal degrees.
    3. longitude: Longitude of the earthquake's epicenter in decimal degrees.
    4. depth: Depth of the earthquake in kilometers.
    5. mag: Magnitude of the earthquake.
    6. magType: Magnitude type used to report the earthquake magnitude.
    7. nst: Total number of seismic stations used to calculate location and magnitude.
    8. gap: Largest azimuthal gap between adjacent stations in degrees.
    9. dmin: Distance to the nearest station in degrees.
    10. rms: Root-mean-square of the residuals of the earthquake's hypocenter location.
    11. place: Human-readable description of the earthquake's location.
    12. type: Type of seismic event (e.g., earthquake, quarry blast, explosion).
    13. horizontalError: Horizontal error in kilometers of the location.
    14. depthError: Depth error in kilometers.
    15. magError: Estimated standard error of the reported earthquake magnitude.
    16. magNst: Number of seismic stations used to calculate the earthquake magnitude.
    17. status: Status of the earthquake event in the USGS earthquake catalog.18. locationSource: ID of the agency or network that provided the earthquake location.
    19. magSource: ID of the agency or network that provided the earthquake magnitude.

# K-Means Clustering


# 1 Determine Value of 'K' in K-Means using Elbow Curve & KMeans-Inertia



```python
'''
KMeans-Inertia : Sum of Squared Distances of Samples to their closest Cluster Center (Centroid), Weighted by the Sample Weights (if provided)
'''
wcssd = [] # Within-Cluster-Sum-Squared-Distance
nr_clus = range(1,11) # Number of Clusters
for k in nr_clus:
    kmeans = kmclus(n_clusters=k, init='random', random_state=111)
    kmeans.fit(earthquake_subset)
    wcssd.append(kmeans.inertia_)
plt.plot(nr_clus, wcssd, marker='x')
plt.xlabel('Values of K')
plt.ylabel('Within Cluster Sum Squared Distance')
plt.title('Elbow Curve for Optimal K')
plt.show()
```
